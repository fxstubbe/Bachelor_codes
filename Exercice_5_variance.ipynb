## Ex5: Calculer la moyenne et la variance de 1000 nombres générés aléatoirement entre 0 et 1. Représenter ces données dans un histogramme. Même consigne pour des données générées par une loi normale de moyenne 0 et écart-type 1.


### 1) Packages

```python
import random
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

### 2) Calcul de la moyenne et de la variance de nombres aléatoires

Homemade fonctions

```python
numbers = [random.random() for i in range(1000)]
moyenne = sum(numbers)/len(numbers)
variance = sum ( [ (i - moyenne)**2 for i in numbers ] ) / ( len(numbers) - 1 )
print("Moyenne : " + str(moyenne))
print("Variance : " + str(variance))
```

```python
sns.set(style = 'whitegrid')
sns.distplot(numbers)
plt.show()
```

on peut aussi exploiter numpy

```python
random_numbers = np.random.random(1000)
output = (np.mean(random_numbers), np.var(random_numbers))
print(output)
```

```python
sns.set(style = 'whitegrid')
sns.distplot(random_numbers)
plt.show()
```

### 3) Depuis des nombres d'une distribution normale

```python
normal_numbers = np.random.normal(size = 1000)
output = (np.mean(normal_numbers), np.var(normal_numbers))
print(output)
```

```python
sns.set(style = 'whitegrid')
sns.distplot(normal_numbers)
plt.show()
```
