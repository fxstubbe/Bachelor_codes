## Ex6: Le module "time" sert à estimer le temps qu'il faut à un code pour s'exécuter. Adapter le code ci-dessous pour estimer le temps nécessaire à l'exécution de 100 et de 1000 additions. Que pouvez-vous observer? Représenter graphiquement le temps d'exécution du code en fonction du nombre d'opérations (échelle logarithmique de 10 à 10^8 opérations)

### 1) Packages

```python
import time
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

### 2) La fonction  

```python
def time_addition(times):
    start_time = time.clock()
    count = 0
    for i in range (times):
        count += 1 
    end_time = time.clock() 
    return(end_time - start_time)

iterations = [10**i for i in range (1,9)]

mydata = {'#Addition': [i for i in iterations], ' Time taken (s) ': [time_addition(i) for i in iterations]}
mydataframe = pd.DataFrame(data=mydata)
```

### 3) Visualisation

```python
sns.set()
sns.lmplot(x = "#Addition", y = ' Time taken (s) ', data = mydataframe)
plt.show()
```
