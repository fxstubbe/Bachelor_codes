### Exercice 1 

Create a function that will replace every vowel of a string of characters by a "X". Example: "hello" -> "hXllX".

```python
def remplace(phrase, rempla):
    string = str(phrase)
    newstring = ''

    for lettre in string:
        if lettre.lower() not in 'aeiou':
            newstring += str(lettre)
        else:
            newstring += str(rempla)
    return newstring
remplace('Ma phrase', 'X')
```

La même fonction en utilisant une compréhension de liste

```python
def vowel_converter(string, replacement):
    return ''.join([letter if letter.lower() not in 'aeiou' else replacement for letter in string])

vowel_converter('hello', 'X')
```

Write a program that can compute the roots of the polynom  y=x2+bx+c . If no solution exists, then write the message "No solution found!".

```python
from math import sqrt
```

```python
a = (2,4,3)
b = (2,4,1)
```

```python
def square_roots(poly):
    delta = poly[1]**2 - 4*poly[0]*poly[2]
    if delta >= 0:
        root1 = (poly[1]+sqrt(delta))/2*poly[0]
        root2 = (poly[1]-sqrt(delta))/2*poly[0]
        return (root1, root2)
    else :
        return "No solution found ! "
print(square_roots(a))   
print(square_roots(b))  
```

## Exercice 6

The library time is useful to estimate how long a code runs. Adapt the code below to estimate how long it would take to perform  10**9  additions on your computer. Does the time scale with the number of operations? Could you plot this time as a function of the number of operations?

```python
import time
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def time_addition(times):
    start_time = time.clock()
    count = 0
    for i in range (times):
        count += 1 
    end_time = time.clock() 
    return(end_time - start_time)

iterations = 10000

mydata = {'#Addition': [i for i in range(iterations)], 'Time taken': [time_addition(i) for i in range(iterations)]}
mydataframe = pd.DataFrame(data=mydata)
```

```python
sns.set()
sns.lmplot(x = "#Addition", y = "Time taken", data = mydataframe)
plt.ylim(0, mydataframe['Time taken'].max())
plt.xlim(0, None)
plt.show()
```

```python
time = mydataframe["Time taken"].tolist()
operations = mydataframe["#Addition"].tolist()
slope, intercept, r_value, p_value, std_err = stats.linregress(operations,time)
```

```python
input_estimation = 10**9
output_estimation = input_estimation*slope+ intercept
print ("Doing 10**9 operations will +/- take " + str(output_estimation) + "s")
```

## Exercice 7

Generate random points on the 2D plane in the square [[0,1],[0,1]]. Based on the location of those points, provide a numerical estimation of  π .

```python
import random 
```

```python
def get_pi(repeats):
    inside = 0
    for repeat in range(repeats):
        x, y = random.random(), random.random()
        if (((0.5 - x) ** 2) + ((0.5 - y) ** 2)) <= 0.25:
            inside += 1
        
    return (4 * inside) / repeats

print(get_pi(10**7))
```

## Exercice 8

Consider a walker moving on a bridge. He starts from the center the bridge and, at each time step, he either walks one meter to the right or one meter to the left. Knowing that the bridge is 7m wide, numerically explore the probability that he falls off the bridge after T steps

```python
import random
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

Paramètres pour la simulation

```python
Steps = 100
Walkers = 100000
```

On génère un dataFrame avec une colonne contenant les pas qu'on va faire et une contenant le nombre de marcheurs qui sont tombés à ce pas précis (0 partout pour commencer)

```python
rWalk = pd.DataFrame({'NB_Steps' : [steps for steps in range(Steps+1)] , 'Falling_Walkers' : [0 for walkers in range(Steps+1)]}, columns=['NB_Steps', 'Falling_Walkers'])
rWalk.shape
```

On définit une fonction qui fait marcher le walker en avant ou en arrière

```python
def Walk():
    deplacement = random.choice([-1,1])
    return deplacement
```

On fait une fonction qui retourne le nombre de pas que le marcheur a fait avant de tomber

```python
def Falling_distance(max_steps):
    position = 0
    for steps in range (max_steps):
        position += Walk()
        if position > 3 or position < -3:
            return steps+1
            break
```

On lance plusieurs marcheurs simultanément et on incrémente dans le dataframe

```python
for i in range (Walkers):
    FallingStep = Falling_distance(Steps)
    if FallingStep == None : 
        pass
    else :
        rWalk.iloc[FallingStep,1] += 1
```

On ajoute une colonne, contenant la probabilité, au dataframe 

```python
col = rWalk.apply ( lambda row : ((row['Falling_Walkers']) / Walkers)*100, axis = 1 ) 
rWalk["% Falling"] = col
```

On graph les probabilités. On observe qu'on ne peut pas tomber aux pas impairs et que la probabilité de tomber diminue. Toutefois, il semblerait logique que la pobabilité de tomber augmente dans les premiers (tomber à 4 signifie qu'on marche tout droit, ce qui semble moins probable que faire quelques pas en avant et quelques pas en arrière avant de tomber).

```python
sns.set()
sns.lmplot('NB_Steps', '% Falling', rWalk, fit_reg = None)
plt.ylim(0, rWalk['% Falling'].max()+2)
plt.xlim(0, None)

plt.show()
```

```python
def Walk():
    deplacement = random.choice([-1,1])
    return deplacement
```

```python
def Falling_distance_2(max_steps, start_position):
    position = start_position
    for steps in range (max_steps):
        position += Walk()
        if position > 3: 
            return 'Right', steps+1
            break
        if position < -3:
            return 'Left', steps+1
            break
        else : 
            return 'Alive', None
Falling_distance_2(Steps, 0)
```

```python
for i in range (Walkers):
    Side, FallingStep = Falling_distance_2(Steps,start)
    if FallingStep == None : 
        pass
    else:
        if Side == 'Left':
            rWalk.iloc[FallingStep, 2] += 1 
        else:
            rWalk.iloc[FallingStep, 3] += 1  
```

```python
col1 = rWalk.apply ( lambda row : ((row['Left']) / Walkers)*100, axis = 1 ) 
col2 = rWalk.apply ( lambda row : ((row['right']) / Walkers)*100, axis = 1 ) 
rWalk["%Falling_Left"] = col1
rWalk["%Falling_Right"] = col2
```

```python
import random
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
Steps = 200
Walkers = 10**5
```

```python
data = pd.DataFrame({'NB_Steps' : [steps for steps in range(Steps+1)] , 'X' : [0 for __ in range(Steps+1)], 'Y' : [0 for __ in range(Steps+1)]}, columns=['NB_Steps', 'X', 'Y'])
```

```python
def Walk(steps):
    x, y = 0,0
    for __ in range (steps):
        dx, dy = random.choice([(0,1), (0,-1), (1,0), (-1,0)])
        x += dx
        y += dy
    return (x,y)

def Average_Position(Steps, Walkers):
    x,y = 0,0
    for __ in range(Walkers):
        dx, dy = Walk(Steps)
        x += dx
        y += dy
    x,y = (x/Walkers)*100, (y/Walkers)*100
    return(x,y)
```

```python
for i in range (Steps):
    x, y = Average_Position(i, 100)
    data.iloc[i,1] += x
    data.iloc[i,2] += y
```

```python
for i in range (Steps):
    x, y = Average_Position(i, 100)
    data.iloc[i,1] += x
    data.iloc[i,2] += y
```

```python
ax = sns.kdeplot(data.X, data.Y, shade = True, cmap = "Blues")
ax.set(xlabel = 'Ending Position X',  ylabel = 'EndingPosition Y')
plt.xlim(-Steps,Steps)
plt.ylim(-Steps, Steps)
plt.show()
```

```python
data = pd.DataFrame({'NB_Steps' : [steps for steps in range(Steps+1)] , 'Center' : [0 for __ in range(Steps+1)]}, columns=['NB_Steps', 'Center'])
```

```python
def walk():
    return random.choice([(0,1), (0,-1), (1,0), (-1,0)])

def back_center(max_steps):
    x, y, steps = 0,0,0
    for __ in range (max_steps):
        dx,dy = walk()
        x += dx
        y += dy
        steps += 1
        if x == 0 and y ==0 :
            return (True, steps)
        if steps == max_steps:
            return (False, max_steps)
back_center(2)
```

```python
for i in range(Walkers):
    center_step = back_center(Steps)
    if center_step[0] is True: 
        data.iloc[center_step[1], 1] += 1
```

```python
col = data.apply ( lambda row : ((row['Center']) / Walkers)*100, axis = 1 ) 
data["%Back_Center"] = col
```
